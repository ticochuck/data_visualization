```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import World Population Data

```python
population = pd.read_csv("data/population_total_long.csv")
```

### Filter data to USA population data only 

```python
year = population['Year']
us_population = population['Country Name'] == 'United States'
costa_rica = population['Country Name'] == 'Costa Rica'
mexico = population['Country Name'] == 'Mexico'
```

```python
us_data = population.loc[us_population, ["Country Name", "Year", "Count"]]
crc_data = population.loc[costa_rica, ["Country Name", "Year", "Count"]]
mex_data = population.loc[mexico, ["Country Name", "Year", "Count"]]
```

### Population growth for United States, Costa Rica, and Mexico by year in millions

```python
plt.plot(us_data['Year'], us_data['Count'] / 10**6)
plt.plot(crc_data['Year'], crc_data['Count'] / 10**6)
plt.plot(mex_data['Year'], mex_data['Count'] / 10**6)
plt.title('Population growth(MM) for US, Costa Rica, and Mexico')
plt.legend(['United States', 'Costa Rica', 'Mexico'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

<p> The grapth above makes look like the population growth for Costa Rica was flat over the years. This is because the difference in total population by country. In the example below, I'm ging to compare the same data, but using a percentage growth <p>

### Calculate population growth by year, by percentage

```python
us_percent = (us_data['Count'] / us_data['Count'].iloc[0]) * 100
crc_percent = (crc_data['Count'] / crc_data['Count'].iloc[0]) * 100
mex_percent = (mex_data['Count'] / mex_data['Count'].iloc[0]) * 100
```

```python
plt.plot(us_data['Year'], us_percent)
plt.plot(crc_data['Year'], crc_percent)
plt.plot(mex_data['Year'], mex_percent)
plt.legend(['United States', 'Costa Rica', 'Mexico'])
plt.xlabel('Year')
plt.ylabel('Population Growth (Starting at 100%)')
plt.show()
```

<p> This view not only allows us to see the growth for Costa Rican population, but it also shows that the population grew more rapitly in Costa Rica, than it did for Mexico, or the United States  <p>
