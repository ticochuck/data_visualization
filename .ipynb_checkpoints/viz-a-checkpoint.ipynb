```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Tips

```python
tips = pd.read_csv('data/tips.csv')
tips
```

<p> One waiter recorded information about each tip he received over a period of a few months working in one restaurant. In all he recorded 244 tips.

```python
vis2 = sns.set_style('whitegrid')
time = sns.FacetGrid(tips, col = "time")
time.map(plt.hist, "tip")
```

```python
color_set = dict(Lunch="gray", Dinner="purple")
vis1 = sns.catplot(x="sex", y="tip", data=tips, hue="time", kind='box', palette=color_set)
```

```python
daily = sns.lineplot(x="day", y="tip", hue="sex", data=tips, ci=0)
```

```python
tips['tip_percentage'] = tips['tip'] / tips['total_bill'] * 100
```

```python
sns.set()
plt.figure()
x2 = tips['tip_percentage']
x3 = tips['tip']
plt.plot(x2, label="Tip %")
plt.plot(x3, label="Tip")
plt.legend()
plt.show()
```

```python
tot_tips = tips['tip_percentage'].mean().round(2)
```

```python
woman = tips['sex'] == "Female"
a = tips.loc[woman, "tip_percentage"].mean().round(2)
```

```python
man = tips['sex'] == "Male"
b = tips.loc[man, "tip_percentage"].mean().round(2)
```

```python
plt.figure()
names= ['Female', 'Male', 'Total Avg']
x1 = [a, b, tot_tips]
pal = ['seagreen', 'gray', 'lightblue']
plt.bar(names, x1, color=pal)
plt.show()
```

```python
tip_percent = sns.lineplot(x="day", y="tip_percentage", hue="sex", data=tips, ci=0)
```

```python
total_bill = sns.lineplot(x="day", y="total_bill", hue="sex", data=tips, ci=0)
```

<p> Based on data above, we know that:
    <li> tips are higher during Dinner Time </li>
    <li> Man typically leave a higher amount tip, except for Sundays </li>
    <li> Woman total bill is usually smaller, which is why their tips are lower than man's</li>
    <li> Woman leaves a higher tip in relation to the total bill </li>
